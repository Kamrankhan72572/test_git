```python
# import library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# load data set
chilla= pd.read_csv("Chilla_data_plots.csv")
sns.set(style="whitegrid")
sns.boxplot(x="Age",
                y="Your Weight in kg? (float)",
                palette='Set2', dodge= True,
                showmeans=True,   # show mean
                 meanprops={"marker":"+",
                             "markersize":"12",
                             "markeredgecolor": "red"}, data=chilla)
# show lables  
plt.xlabel("how much are fat", size="24", weight="bold"),
plt.ylabel("age(years)" ,size="24", weight="bold"),
plt.title("weak and fat" ,size="24", weight="bold")

                 
            
```
