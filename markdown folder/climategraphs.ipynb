```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
climate_change=pd.read_csv("Global_Tempraturechange_1960_2020.csv")
climate_change
```

```python
#draw line plot
sns.lineplot(x="Year",y="Value", hue="Element", data=climate_change)
plt.title("Global_Climate_Change")
sns.set_style("dark")
plt.show()
```
