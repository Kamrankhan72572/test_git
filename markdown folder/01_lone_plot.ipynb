```python
#step1/ import libraries
import seaborn as sns
import matplotlib.pyplot as plt
# load data set
phool=sns.load_dataset("iris")
# size figure
plt.Figure(figsize=(8,6))
#draw line plot
sns.lineplot(x="sepal_length",y="sepal_width", data=phool)
plt.title("flower plot")
sns.set_style("dark")
plt.show()

```
