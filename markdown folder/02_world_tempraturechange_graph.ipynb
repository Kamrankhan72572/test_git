```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
climate_change= pd.read_csv("Global_Tempraturechange.csv")
plt.Figure(figsize=(8,6))
sns.barplot(x="Year",y="Value", hue="Element", data=climate_change)
plt.title("Global climate change")
sns.set_style("dark")
plt.show()
```
